### Merging all cleaned dataset across different platforms

```python
import pandas as pd
```

#### TV Shows dataset merging

```python
#reading in the tv show files and adding them to one data frame together
amazon_tv = pd.read_csv("./Data_cleaning/AmazonTvShows.csv")
hulu_tv = pd.read_csv('./Data_cleaning/Hulu_tv_shows.csv')
netflix_tv = pd.read_csv('Data_cleaning/netflix_tv.csv')
disneyplus_tv = pd.read_csv('Data_cleaning/disneyplus_tv.csv')
amazon_tv
```
